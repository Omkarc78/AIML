```python
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,8])
ypoints=np.array([3,10])
#z=np.array([4,9])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
x=range(1,50)
y=[value*3 for value in x]
print("Value of x: ")
print(*range(1,50))
print("Value in Y(thrice of X):")
print(y)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.xlabel('y-axis')
plt.title('Draw a line')
plt.show()
```

```python
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
x=[1,2,3]
#x axis value
y=[2,4,1]
#y axis
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x=np.array([1,1,6,6,1,6,1,6])
y=np.array([3,8,8,3,3,8,8,3])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,'o-.b')
plt.show()
```

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='*')
plt.show
```

```python
y=np.array([3,8,1,10])
plt.plot(y,'o:r')
plt.show
```

```python
y=np.array([3,8,1,10])
plt.plot(y,'o-.m')
#plt.show
```

```python
y=np.array([3,8,1,10])
plt.plot(y,linestyle='dashed',linewidth=10.1)
# plt.show()
```
